# Week 4 Regression Engine Demo

Demonstrates one LSM time-slice regression with ITM filtering and discounted realized continuation targets.


```python
import numpy as np
from lsm.simulation import simulate_gbm_paths
from lsm.regression import estimate_continuation_value
```

```python
s0, k, r, sigma, t = 40.0, 40.0, 0.06, 0.2, 1.0
n_paths, n_steps = 50_000, 50
paths = simulate_gbm_paths(s0=s0, r=r, sigma=sigma, maturity=t, n_paths=n_paths, n_steps=n_steps, seed=21)
time_index = 25
stock_t = paths[:, time_index]
# Week 4 proxy target (demo): terminal put payoff. Week 5 replaces this with recursively updated future cash flows.
future_cf = np.maximum(k - paths[:, -1], 0.0)
result = estimate_continuation_value(stock_prices_t=stock_t, realized_future_cashflows=future_cf, strike=k, rate=r, dt=t/n_steps)
result.status, result.n_itm, result.coefficients
```
