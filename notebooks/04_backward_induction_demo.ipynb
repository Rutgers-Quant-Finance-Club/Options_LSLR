# Week 5 Backward Induction Demo

Fits an LSM policy in-sample and applies it out-of-sample.


```python
from lsm.simulation import simulate_gbm_paths
from lsm.american import price_american_put_lsm, price_american_put_with_policy
```

```python
s0, k, r, sigma, t = 40.0, 40.0, 0.06, 0.2, 1.0
train = simulate_gbm_paths(s0, r, sigma, t, n_paths=80_000, n_steps=50, seed=101)
test = simulate_gbm_paths(s0, r, sigma, t, n_paths=80_000, n_steps=50, seed=202)
fit = price_american_put_lsm(paths=train, strike=k, rate=r, maturity=t)
out = price_american_put_with_policy(paths=test, policy=fit.policy)
fit.price, fit.std_error, out.price, out.std_error
```
